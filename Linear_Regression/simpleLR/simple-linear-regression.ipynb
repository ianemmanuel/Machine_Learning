#### PREDICTING HOUSE PRICES BASED ON AREA (SQUARE FT)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
home_prices_df = pd.read_csv('./homeprices.csv')
```

```python
home_prices_df
```

```python
plt.title('HOUSE PRICE SCATTER PLOT')
plt.xlabel('Area (Sqft)')
plt.ylabel('Price (USD)')
plt.scatter(home_prices_df['area'],home_prices_df['price'])
```

```python
reg_model = linear_model.LinearRegression()
reg_model.fit(home_prices_df[['area']],home_prices_df.price)
```

```python
reg_model.coef_
```

```python
reg_model.intercept_
```

```python
plt.title('HOUSE PRICE PREDICTION')
plt.xlabel('Area (Sqft)')
plt.ylabel('Price (USD)')
plt.scatter(home_prices_df['area'],home_prices_df['price'])
plt.plot(home_prices_df['area'], reg_model.predict(home_prices_df[['area']]))
```

```python
predict_df = pd.read_csv('./price_prediction.csv')
```

```python
predict_df 
```

```python
reg_model.predict(predict_df)
```

```python
predict_df['predicted_prices'] = reg_model.predict(predict_df)
```
